```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
app_df = pd.read_csv('Resources/googleplaystore.csv')
app_df.head()
```

```python
app_df_clean = app_df.dropna(how='any')
```

```python
app_df_new  = app_df_clean[['App','Rating','Category','Reviews','Size','Installs','Type','Content Rating']]
app_df_new.head()
```

```python
app_df_new = app_df_new.astype({'Reviews':int})
```

```python
app_df_new = app_df_new[app_df_new.Size != 'Varies with device']
app_df_new['Size'] =app_df_new['Size'].replace({'k': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(int)
```

```python
app_df_new['Installs'] = app_df_new['Installs'].apply(lambda x : x.strip('+').replace(',', ''))
app_df_new['Installs'] = app_df_new.Installs.astype(int)
```

```python
app_df_new.head()
```

```python
rating_count = app_df_new.Rating.value_counts()
```

<h4>WoE Encoding of App Categories

```python
#Category, total Not Category
category_count = app_df_new.Category.value_counts()
category_df = pd.DataFrame(category_count)

category_total = category_df.Category.sum()

category_df['Not Category'] = category_total-category_df['Category']
category_df['WoE'] = np.log(category_df['Category']/category_df['Not Category'])
category_df['IV'] = np.log(category_df['WoE'] *(category_df['Category'] -category_df['Not Category']))
category_df.head()
```

<h4>WoE Encoding of App Types

```python
#Type, Total Not Type
type_count = app_df_new.Type.value_counts()
type_df = pd.DataFrame(type_count)

type_total = type_df.Type.sum()

type_df['Not Type'] = type_total-type_df['Type']
type_df['WoE'] = np.log(type_df['Type']/type_df['Not Type'])
type_df['IV'] = np.log(type_df['WoE'] *(type_df['Type'] -type_df['Not Type']))
type_df
```

<h4>WoE Encoding of App Content Rating

```python
#Content Rating, Total Not Content Rating 
content_count = app_df_new['Content Rating'].value_counts()
content_df = pd.DataFrame(content_count)

content_total = content_df['Content Rating'].sum()

content_df['Not Content Rating'] = content_total-content_df['Content Rating']
content_df['WoE'] = np.log(content_df['Content Rating']/content_df['Not Content Rating'])
content_df['IV'] = np.log(content_df['WoE'] *(content_df['Content Rating'] -content_df['Not Content Rating']))
content_df
```

<h4>Mapping WoE encoded variables to DataFrame

```python
woe_encoded_data = app_df_new
woe_encoded_data.loc[:,'Category_WoE'] = app_df_new['Category'].map(category_df['WoE'])
woe_encoded_data.loc[:,'Type_WoE'] = app_df_new['Type'].map(type_df['WoE'])
woe_encoded_data.loc[:,'ContentRating_WoE'] = app_df_new['Content Rating'].map(content_df['WoE'])

woe_encoded_data.tail(20)
```

```python
woe_encoded_data = woe_encoded_data.drop(columns=['Category', 'Type', 'Content Rating'])
```

```python
woe_encoded_data.tail(20)
```

<h4>RFR model evaluation 

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    errors = abs(y_pred-y_test)
    mape=100*np.mean(errors/y_test)
    accuracy=100-mape
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    print('MAE: ', mean_absolute_error(y_test, y_pred))
    print('MSE: ', mean_squared_error(y_test, y_pred)) 
    print('RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred)))
    return accuracy
```

```python
X = woe_encoded_data.iloc[:,2:]
y = woe_encoded_data.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100).fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
base_accuracy_rf = evaluate(rf, X_test, y_test)
```

Results of similiar accuracy, error standards as base Random Forest Regression model using one-hot encoded variables.<br>
WoE is typically used in optimization of Logistic Regression model with binary target. 

```python
f,ax=plt.subplots(1,2,figsize=(18,6))

feat_sorted_idx = rf.feature_importances_.argsort()
ax[0].bar(X_train.columns[feat_sorted_idx ], rf.feature_importances_[feat_sorted_idx ])
ax[0].set_title("Feature Importance")

from sklearn.inspection import permutation_importance
perm_importance = permutation_importance(rf, X_test, y_test)
perm_sorted_idx = perm_importance.importances_mean.argsort()
ax[1].bar(X_test.columns[perm_sorted_idx], perm_importance.importances_mean[perm_sorted_idx])
ax[1].set_title("Permutation Importance")
```

```python
shap.summary_plot(shap_values, X_test, plot_type="bar")

shap.summary_plot(shap_values, X_test)
```
